```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

from datetime import time
from datetime import datetime
from datetime import timedelta

plt.rcParams['font.family'] = 'Palatino Linotype'
plt.rcParams['font.size'] = 12
```

```python
df_voltage = pd.read_excel('./article/DATA.xlsx', sheet_name="Sheet2")
df_voltage.Time = pd.to_datetime(df_voltage.Time, format='%H:%M:%S')
df_voltage.Time = (df_voltage.Time - df_voltage.Time[0]).dt.total_seconds() / 60

print(df_voltage.info())
df_voltage
```

```python
# plot search of the optimal parameter n_components of PCA
plt.figure(figsize=(16, 3))
plt.subplot(1, 2, 2)

plt.plot(df_voltage.Time, df_voltage.Voltage, '#1f77b4') # , label='voltage')
plt.xlabel('Time, min')
plt.xlim(0, 500)
plt.ylabel('Voltage, V')
plt.ylim(30, 55)
plt.legend('Voltage')
plt.title('The SOFC system output electrical characteristics')

plt.annotate('Work under load', xy=(225, 39.5), 
             xycoords='data', xytext=(140, 33), 
             textcoords='data', fontsize=12, 
             arrowprops=dict(arrowstyle='-|>'))

plt.annotate('', xy=(340, 33), 
             xycoords='data', xytext=(230, 33.3), 
             textcoords='data', fontsize=12, 
             arrowprops=dict(arrowstyle='-|>'))

plt.annotate('Battery powered', xy=(100, 52), 
             xycoords='data', xytext=(50, 45), 
             textcoords='data', fontsize=12, 
             arrowprops=dict(arrowstyle='-|>'))

plt.annotate('', xy=(400, 52), 
             xycoords='data', xytext=(140, 45.5), 
             textcoords='data', fontsize=12, 
             arrowprops=dict(arrowstyle='-|>'))

plt.tight_layout()
plt.savefig('pictures/The SOFC system output electrical characteristics.png', dpi = 300, transparent = True)
plt.show()
```
